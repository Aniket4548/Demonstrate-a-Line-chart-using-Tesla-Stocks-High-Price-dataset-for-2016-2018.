# Demonstrate a Line chart using Tesla Stocks High Price dataset for 2016-2018.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#line Chart
stock_data = pd.read_csv("D:/Machine Learning/tesla-stock-price.csv")
```

```python
stock_data.head()
```

```python
stock_data['date'] = pd.to_datetime(stock_data['date'])
stock_data.head()
```

```python
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_axes([0,0,1,1])
ax1.set_title('Teasla Stocks High Stock Prices')
ax1.plot(stock_data['date'],stock_data['high'])
plt.xticks(rotation=90)
plt.show()
```
